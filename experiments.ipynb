```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb
import pickle 
from sklearn.preprocessing import StandardScaler , LabelEncoder
```

```python
churn=pd.read_csv('Churn_Modelling.csv')
```

```python
churn.head()
```

```python
churn.info()
```

```python
#here we can see there is object which we have to change into the integer value and also no null value it is good thing 
#there are many column which did not effect on the output so, need to remove that also
```

```python
churn=churn.drop(['RowNumber' , 'CustomerId' ,'Surname'] ,axis=1)
```

```python
churn
```

```python
#now after removing unnecessary columns need to do encoding and standardization .
#now my will do labelEncoding 

from sklearn.preprocessing import LabelEncoder
label_encoder=LabelEncoder()
churn['Gender']=label_encoder.fit_transform(churn['Gender'])
churn.head(10)
```

```python
#Since we have gender in binary form means only two option then we use lableEncoder but for the Geography we ahve toatl theree types of data then for this we need to use differne method for example one hot encodeing

from sklearn.preprocessing import OneHotEncoder
one_hot=OneHotEncoder()
onehotencoder_geo=one_hot.fit_transform(churn[['Geography']]) 
#need to give 2d array 
#check the datatype of onehotencoder_geo
onehotencoder_geo
```

```python
#change it into array 
onehotencoder_geo_array=onehotencoder_geo.toarray()
```

```python
#creting a dataframe with the columns name as defined by the onehotencoder_geo
onehotencoder_featurename=one_hot.get_feature_names_out(['Geography'])
```

```python
# now create a datafraame so, we can easily add this to the main grp 
onehotencoder_geo_dataframe=pd.DataFrame(onehotencoder_geo_array ,columns=onehotencoder_featurename)
```

```python
onehotencoder_geo_dataframe.head()
```

```python
# now concatinating both dataframe
updated_churn=pd.concat([churn , onehotencoder_geo_dataframe], axis=1)
```

```python
updated_churn.head()
```

```python
# now save the file so, we can use this later
#saving first labelencoder
with open('label_encoder_gender.pkl','wb') as file:
    pickle.dump(label_encoder, file)
```

```python
# now saving the onehotencoder file 
with open('one_hot_encoder_geo.pkl','wb') as file:
    pickle.dump(one_hot, file)
```

```python
#Since we don't need it so, remove geography from the dataframe
updated_churn=updated_churn.drop(['Geography'] ,axis=1)
```

```python
# Break data into independent and dependent features
X=updated_churn.iloc[:,:-1]
X
```
